### Entrada:
#### Para explicar como funcionan las entradas, veremos un ejemplo acerca de la condici贸n para la categorizaci贸n: Coeficiente de variaci贸n [CV]
* Si CV del bloque es < 25% : Medido
* Si CV del bloque es >= 25% y <40% : Indicado
* Si CV del bloque es >= 40%: Inferido

**CV [%] = ( / 渭)*100**

```python
from math import sqrt
```

```python
z=float(input("Ingrese el valor estimado del bloque [Z*(x)]: "))
var=float(input("Ingrese la varianza de estimaci贸n del bloque [var^2(x)]: "))

resultado=((sqrt(var))/z)*100
print("CV del bloque: ", resultado, "%")
if resultado < 25:
    print("Bloque medido")
elif resultado >=25:
    print("Bloque indicado")
elif resultado >=40:
    print("Bloque inferido")
```
